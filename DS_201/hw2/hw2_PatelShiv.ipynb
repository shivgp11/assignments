```python
import pandas as pd
import seaborn as sb
```

```python
df = pd.read_csv ('Lemonade2016-2.csv')
df.head()
df.set_value(8, "Date", '7/8/2016')
df.drop_duplicates(inplace = True)
df.insert(7, "Sales", 0)
df['Sales'] = (df['Orange'] + df['Lemon'])
df.insert(8, "Revenue", 0)
df['Revenue'] = (df['Sales'] * df['Price'])
ax1 = df.plot.scatter(x='Leaflets',y='Sales',c='DarkBlue')
df.hist(column = ['Revenue'])
df.boxplot(by ='Location', column =['Sales'], grid = False)
df2 = df[df['Temperature'] > 74] 
df2.boxplot(by ='Location', column =['Sales'], grid = False)
df = df.drop(columns = ['Date', 'Price', 'Lemon', 'Orange', 'Sales'])
```

```python
sb.pairplot(df, hue = 'Location', diag_kind = 'hist')
print(sb)
```
